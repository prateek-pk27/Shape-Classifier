```python
import numpy as np
import pandas as pd
import glob
import cv2
from keras.preprocessing.image import img_to_array, load_img
from keras.models import Model
from keras.layers import Input, Dense, Flatten, Dropout
from keras.layers.convolutional import Conv2D
from keras.layers.pooling import MaxPooling2D
from keras.layers.normalization import BatchNormalization
from skimage.color import rgb2gray
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python
category_list = sorted(glob.glob(root_path + "./Input/Dataset/train/*"))
category_list
```

```python
X_train = []
Y_train = []
for category in category_list:
    imgs = glob.glob(category+"/*.png")
    for img in imgs:
        img_array = cv2.resize(img_to_array(load_img(img)), dsize=(400, 400), interpolation=cv2.INTER_LINEAR)
        X_train.append(img_array)
        Y_train.append(int(category.split(' ')[-1])-1)
```

```python
X_train = np.array(X_train)
X_train.shape
```

```python
Y_train = to_categorical(Y_train, num_classes=4)
```

```python
inputs = Input(shape=X_train.shape[1:])

conv1 = Conv2D(16, kernel_size=3, activation='relu', padding = 'same', strides = 2)(inputs)
norm1 = BatchNormalization()(conv1)
pool1 = MaxPooling2D(pool_size=(3,3))(norm1)
drop1 = Dropout(0.6)(pool1)

conv2 = Conv2D(32, kernel_size=3, activation='relu', padding = 'same')(drop1)
norm2 = BatchNormalization()(conv2)
pool2 = MaxPooling2D(pool_size=(3,3))(norm2)
drop2 = Dropout(0.5)(pool2)
flat2 = Flatten()(drop2)

hidden1 = Dense(30, activation='relu')(flat2)
hidden2 = Dense(10, activation='relu')(hidden1)

output = Dense(4, activation='softmax')(hidden2)

model = Model(inputs = inputs, outputs = output)
```

```python
 x_train, x_test, y_train, y_test = train_test_split(X_train, Y_train, test_size=0.1, random_state=100)
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
history = model.fit(x_train, y_train, epochs=50, batch_size=32, verbose=1, shuffle = True, validation_data=(x_test, y_test))
```

```python
test_list = sorted(glob.glob(root_path + "./Input/Dataset/test/*"))
test_list[0:5]
```

```python
Y_test = []
for test in test_list:
    img_array = cv2.resize(img_to_array(load_img(test)), dsize=(400, 400), interpolation=cv2.INTER_LINEAR)
    Y_test.append(img_array)
```

```python
Y_test = np.array(Y_test)
```

```python
pred = model.predict(Y_test)
```

```python
y = (pred.argmax(axis=-1) + 1).tolist()
```

```python
print(y)
```

```python
test_image_id = []
for img in test_list:
    test_image_id.append(img.split('/')[-1])
    
for i in range(len(y)):
    y[i] = "Category " + str(y[i])
    
data = list(zip(test_image_id, y))
print(data)
```

```python
df = pd.DataFrame(data, columns = ['ImageId', 'predicted_labels'])
df.head()
```
